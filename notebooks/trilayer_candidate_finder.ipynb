```python
results = run_and_filter(a1=lattice_vectors1['a'][0:2],a2=lattice_vectors1['b'][0:2],
                      g1=lattice_vectors2['a'][0:2],g2=lattice_vectors2['b'][0:2],
                      degmin=1, degmax = 60.01, degstep = 0.01,
                      dtol = 5e-4, nmin = -10, nmax = 10, goodangles = [60.0])
```

```python
len(results[0]['results'])
```

```python
angles= []
for t in results[0]['results']:
    angles.append(t[0]['angle'])
    
```

```python
unique_angles = sorted(set(angles))

len(unique_angles)
```

```python
angle=[]
A1 = []
A2 = []
delvec = []
delgood = []
norm_A1 = []
norm_A2 = []

for result in results[0]['results']:
    angle.append(result[0]['angle'])
    A1.append(result[0]['matchpoint'])
    A2.append(result[1]['matchpoint'])
    delvec.append(result[0]['delvec'])
    delgood.append(np.abs(result[2]['delgoodangle']))
    norm_A1.append(np.linalg.norm(np.array(result[0]['matchpoint'])))
    norm_A2.append(np.linalg.norm(np.array(result[1]['matchpoint'])))
```

```python
resdf = pd.DataFrame(zip(angle,A1,A2,delvec,delgood,norm_A1, norm_A2), columns=['Theta','A1', 'A2', 'delvec', 'delang','norm_A1', 'norm_A2'])
```

```python
len(resdf)
```

```python
resdf.sort_values(by='delvec', ascending=False)
```

```python
tol = 0.05  # Theta‐difference tolerance for “closeness”

# NEW: filter where norm_A1 ≈ norm_A2 within 0.1
df_filtered = resdf[abs(resdf['norm_A1'] - resdf['norm_A2']) <= 0.1]

# 1) sort by Theta
df_sorted = df_filtered.sort_values('Theta').reset_index(drop=True)


keep_idxs = []           # indices of the rows we'll keep
current_group = [0]      # start with the first row in group

for i in range(1, len(df_sorted)):
    θ_prev = df_sorted.loc[i-1, 'Theta']
    θ_curr = df_sorted.loc[i,   'Theta']

    if abs(θ_curr - θ_prev) <= tol:
        # same cluster: add to current_group
        current_group.append(i)
    else:
        # new cluster starts: pick best from the old
        best = df_sorted.loc[current_group, 'delvec'].idxmin()
        keep_idxs.append(best)
        current_group = [i]

# final cluster
best = df_sorted.loc[current_group, 'delvec'].idxmin()
keep_idxs.append(best)

# select and restore original order (optional)
result = df_sorted.loc[keep_idxs].sort_index().reset_index(drop=True)
```

```python
deg=[]
theta=[]
A1 = []
A2 = []
delvec = []
delgood = []
norm_A1 = []
norm_A2 = []


for key in trilayer.keys():
    print(key)
    degree = key
    for result in trilayer[key][0]['results']:
        deg.append(degree)
        theta.append(result[0]['angle'])
        A1.append(result[0]['matchpoint'])
        A2.append(result[1]['matchpoint']) 
        delvec.append(result[0]['delvec'])
        delgood.append(np.abs(result[2]['delgoodangle']))
        norm_A1.append(np.linalg.norm(np.array(result[0]['matchpoint'])))
        norm_A2.append(np.linalg.norm(np.array(result[1]['matchpoint'])))
```

```python
resdf = pd.DataFrame(zip(deg,theta,A1,A2,delvec,delgood, norm_A1, norm_A2), columns=['Theta','Phi','A1', 'A2', 'delvec', 'delang', 'norm_A1', 'norm_A2'])
```

```python
tol_theta = 0.05  # tolerance for Theta closeness
tol_phi   = 0.05  # tolerance for Phi closeness (adjust as needed)

# 1) Filter where norm_A1 ≈ norm_A2
df_filtered = resdf[abs(resdf['norm_A1'] - resdf['norm_A2']) <= 0.1].copy()

# 2) Sort by Theta, then Phi
df_sorted = df_filtered.sort_values(['Theta', 'Phi']).reset_index(drop=True)

# 3) Assign cluster labels using a simple sequential sweep
#    Two rows are in the same cluster if they are close in BOTH Theta and Phi
cluster_id = 0
cluster_labels = [0] * len(df_sorted)

for i in range(1, len(df_sorted)):
    θ_prev, φ_prev = df_sorted.loc[i-1, 'Theta'], df_sorted.loc[i-1, 'Phi']
    θ_curr, φ_curr = df_sorted.loc[i,   'Theta'], df_sorted.loc[i,   'Phi']

    if abs(θ_curr - θ_prev) <= tol_theta and abs(φ_curr - φ_prev) <= tol_phi:
        cluster_labels[i] = cluster_id
    else:
        cluster_id += 1
        cluster_labels[i] = cluster_id

df_sorted['_cluster'] = cluster_labels

# 4) Within each cluster, pick the row with smallest norm_A1, 
#    breaking ties by smallest delvec
result = (
    df_sorted
    .sort_values(['norm_A1', 'delvec'])
    .groupby('_cluster', sort=False)
    .first()
    .reset_index(drop=True)   # this already drops _cluster (it was the index)
)
```

```python
final = result[(result['Theta']<59) & (result['Phi']<59) & (result['norm_A1']<50)]
```
